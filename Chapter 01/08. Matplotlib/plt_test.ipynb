```python
import matplotlib.pyplot as plt
import numpy as np
x1 =np.array([1, 6, 8, 14, 20,10,])
y1 =np.array([3, 7, 13, 25, 5, 13])


front1 ={'family': 'serif', 'color':'blue', 'size': 20}
front2 = {'family': 'serif', 'color': 'red', 'size': 10}

plt.title('sports watch data', fontdict= front1, loc = 'right')
plt.xlabel('Average Pulse', fontdict = front2, )
plt.ylabel('Calorie Burnage', fontdict = front2)


plt.plot(x1, y1)
plt.grid()
plt.legend( bbox_to_anchor=(1.0, 1.0))
plt.show()
```
